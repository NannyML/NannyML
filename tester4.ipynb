```python
import nannyml as nml
```

```python
reference_df = nml.load_synthetic_car_loan_dataset()[0]
analysis_df = nml.load_synthetic_car_loan_dataset()[1]

#rename repaid column to y_true
reference_df = reference_df.rename(columns={'repaid': 'y_true'})

# map the values of the y_true and y_pred columns from 0 and 1 to "true" and "false"
reference_df['y_true'] = reference_df['y_true'].map({0: 'false', 1: 'true'})
reference_df['y_pred'] = reference_df['y_pred'].map({0: 'false', 1: 'true'})
analysis_df['y_pred'] = analysis_df['y_pred'].map({0: 'false', 1: 'true'})

# rename y_pred_proba to y_pred_proba_true
reference_df = reference_df.rename(columns={'y_pred_proba': 'y_pred_proba_true'})
analysis_df = analysis_df.rename(columns={'y_pred_proba': 'y_pred_proba_true'})

#make another column called y_pred_proba_false and set it to 1 - y_pred_proba_true
reference_df['y_pred_proba_false'] = 1 - reference_df['y_pred_proba_true']
analysis_df['y_pred_proba_false'] = 1 - analysis_df['y_pred_proba_true']

reference_df
```

```python
analysis_df
```

```python
estimator = nml.CBPE(
    y_pred_proba={
        'true': 'y_pred_proba_true',
        'false': 'y_pred_proba_false'},
    y_pred='y_pred',
    y_true='y_true',
    timestamp_column_name='timestamp',
    problem_type='classification_multiclass',
    metrics=['f1'],
    chunk_size=6000,
)
```

```python
estimator.fit(reference_df)
```
